```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#initial conditions
tau = 1.0   # ثبات زمانی
N_A0 = 100  # شرایط اولیه N_A
N_B0 = 0    # شرایط اولیه N_B
t_total = 20  #زمان کل 
dt = 0.1    # گام زمانی

time = np.arange(0, t_total, dt)

#  ایجاد آرایه برای ذخیره تعداد هسته ها و مقداردهی اولیه
N_A = np.zeros_like(time)
N_B = np.zeros_like(time)
N_A[0] = N_A0
N_B[0] = N_B0

# حل عددی با استفاده از روش اویلر
for i in range(1, len(time)):
    N_A[i] = N_A[i-1] + ((N_B[i-1] - N_A[i-1]) /tau) * dt
    N_B[i] = N_B[i-1] + ((N_A[i-1] - N_B[i-1]) / tau) * dt

# حل تحلیلی
#N_A_analytical = (N_A0 + N_B0) / 2 + (N_A0 - N_B0) / 2  * np.exp(-t / tau)
#N_B_analytical = (N_A0 + N_B0) / 2 - (N_A0 - N_B0) / 2 * np.exp(-t / tau)


def equations(t, y):
    NA, NB = y
    dNA_dt = (NB / tau) - (NA / tau)
    dNB_dt = (NA / tau) - (NB / tau)
    return [dNA_dt, dNB_dt]
    
NA0 = 100  # جمعیت اولیه هسته A
NB0 = 0    # جمعیت اولیه هسته B
y0 = [NA0, NB0]
# زمان
t_span = (0, 10)  # بازه زمانی
t_eval = np.linspace(t_span[0], t_span[1], 100)

# حل معادلات دیفرانسیل
solution = solve_ivp(equations, t_span, y0, t_eval=t_eval)

# نتایج
NA = solution.y[0]
NB = solution.y[1]
t = solution.t
# رسم نمودار
plt.figure(figsize=(10, 6))
plt.plot(t, NA, label='NA(t)')
plt.plot(t, NB, label='NB(t)')

plt.plot(time, N_A, label='N_A (Numeric)', color='blue')
plt.plot(time, N_B, label='N_B (Numeric)', color='red')
#plt.plot(time, N_A_analytical, label='N_A (Analytic)', linestyle='dashed', color='cyan')
#plt.plot(time, N_B_analytical, label='N_B (Analytic)', linestyle='dashed', color='orange')
plt.title('Radiactive Decay of Premises A and B')
plt.xlabel('Time (s)')
plt.ylabel('Number of Nuclei (NA, NB)')
plt.legend()
plt.grid()
plt.xlim(0, t_total)
plt.ylim(-10, 120)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#initial conditions
tau = 1.0   # ثبات زمانی
N_A0 = 100  # شرایط اولیه N_A
N_B0 = 0    # شرایط اولیه N_B
t_total = 20  #زمان کل 
dt = 0.1    # گام زمانی

time = np.arange(0, t_total, dt)

#  ایجاد آرایه برای ذخیره تعداد هسته ها و مقداردهی اولیه
N_A = np.zeros_like(time)
N_B = np.zeros_like(time)
N_A[0] = N_A0
N_B[0] = N_B0

# حل عددی با استفاده از روش اویلر
for i in range(1, len(time)):
    N_A[i] = N_A[i-1] + ((N_B[i-1] - N_A[i-1]) /tau) * dt
    N_B[i] = N_B[i-1] + ((N_A[i-1] - N_B[i-1]) / tau) * dt

# حل تحلیلی
N_A_analytical = (N_A0 + N_B0) / 2 + (N_A0 - N_B0) / 2  * np.exp(-time / tau)
N_B_analytical = (N_A0 + N_B0) / 2 - (N_A0 - N_B0) / 2 * np.exp(-time / tau)

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.plot(time, N_A, label='N_A (Numeric)', color='blue')
plt.plot(time, N_B, label='N_B (Numeric)', color='red')
plt.plot(time, N_A_analytical, label='N_A (Analytic)', linestyle='dashed', color='cyan')
plt.plot(time, N_B_analytical, label='N_B (Analytic)', linestyle='dashed', color='orange')
plt.title('Radiactive Decay of Premises A and B')
plt.xlabel('Time (s)')
plt.ylabel('Number of Nuclei (NA, NB)')
plt.legend()
plt.grid()
plt.xlim(0, t_total)
plt.ylim(-10, 120)
plt.show()
```
